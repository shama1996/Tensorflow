```python
search=['laptops', 'Phones', 'Headphones', 'smart watches', 'Professional Cameras', 'Printers', 'monitors', 'Home theater', 'router']
```

```python
ProlinK=[]
Rating=[]
Review=[]
for m in search:
    search = driver.find_element_by_xpath("//div[@class='_3OO5Xc']/input")
    search.clear()
    search.send_keys(m)
    search_button=driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
    search_button.click() #clicking on the search button
    time.sleep(5)
    for j in range(0,19):
        hlink=driver.find_elements_by_xpath("//a[@class='_1fQZEK']")
        for i in hlink:
            h=i.get_attribute('href')
            ProlinK.append(h)
        nxt_button=driver.find_element_by_xpath("//a[@class='_1LKTO3']/span")
        if nxt_button.text=='Next':
            nxt_button.click() 
```

```python
ab=[]
for a in ProlinK:    
    driver.get(a)
    try:
        xt=driver.find_element_by_xpath("//div[@class='col JOpGWq']/a")
        tt=xt.get_attribute("href")
        ab.append(tt)
             
    except(NoSuchElementException):
        pass
```

```python
for i in ab:    
    driver.get(i)
    try:
        a=driver.find_elements_by_xpath("//div[@class='col']/div/div[1]/div") #searching for rating path
        for j in a[2:]:
            Rating.append(j.text) #appending the elements in Rating list

        a=driver.find_elements_by_xpath("//div[@class='t-ZTKy']/div/div") #searching for review path
        for k in a:
            Review.append(k.text)
    except(NoSuchElementException):
        pass
```

```python
for m in productname:
    search = driver. find_element_by_xpath("//input[@id = 'twotabsearchtextbox']")
    search.clear()
    search.send_keys(m)
    search_btn = driver.find_element_by_xpath("//div[@class = 'nav-search-submit nav-sprite']/span/input")
    search_btn.click()
    for j in range(0,19):
        hlink=driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a")
        for i in hlink:
            h=i.get_attribute('href')
            prolink.append(h)
        nxt=driver.find_element_by_xpath("//li[@class='a-last']//a")
        nt=nxt.get_attribute("href")
        driver.get(nt)
```

```python
from tqdm import tqdm
for a in tqdm(temp[::-1]):    
    driver.get(a)    
    
    WebDriverWait(driver,50)
    try:
        xt=driver.find_element_by_xpath("//div[@class= 'a-row a-spacing-medium']/a")
        if xt != None:
            tt=xt.get_attribute("href")
            driver.get(tt)
            xt1=driver.find_elements_by_xpath("//div//a[@class= 'a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold']")       
            if xt1 != None:
                for i in xt1:
                    h=i.get_attribute('href')
                    links.append(h)  
    except(NoSuchElementException):
        pass
```
